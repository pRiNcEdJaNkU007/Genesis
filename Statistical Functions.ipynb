```python
# Automating the calculations for the model() class
class model():
    def __init__(self, ylist, xlist):
        self._xlist = xlist
        self._ylist = ylist
        self._update_stats()

    def _update_stats(self):
        sumxy = sum([x * y for x, y in zip(self._xlist.value, self._ylist.value)])
        n = self._xlist._n
        xbar = self._xlist._average
        ybar = self._ylist._average
        sumx2 = self._xlist._sum2
        sumy2 = self._ylist._sum2
        sxy = sumxy - n * xbar * ybar
        sxx = sumx2 - n * xbar ** 2
        syy = sumy2 - n * ybar ** 2
        self.slope = sxy / sxx
        self.intercept = ybar - self.slope * xbar
        self.covariance = sxy / (n - 1)
        self.correlation = sxy / (sxx * syy) ** 0.5

    def _get_xlist(self):
        return self._xlist

    def _set_xlist(self, xlist):
        self._xlist = xlist
        self._update_stats()

    def _del_xlist(self):
        del self._xlist
        del self.covariance
        del self.intercept
        del self.correlation
        del self.slope

    def _get_ylist(self):
        return self._ylist

    def _set_ylist(self, ylist):
        self._ylist = ylist
        self._update_stats()

    def _del_ylist(self):
        del self._ylist
        del self.covariance
        del self.intercept
        del self.correlation
        del self.slope

    xlist = property(fget=_get_xlist, fset=_set_xlist, fdel=_del_xlist, doc="The changes are updated.")
    ylist = property(fget=_get_ylist, fset=_set_ylist, fdel=_del_ylist, doc="The changes are updated.")
    
# Repeat point 4 with the new attributes xlist and ylist
yVar = variable(1, 2, 3, 4, 5)
xVar = variable(1, 0, 2, 5, 6)
mymodel = model(yVar, xVar)
print(mymodel.correlation)

mymodel.xlist = variable(4, 1, 3, 5, 6)
print(mymodel.correlation)

mymodel.ylist = variable(-1, -2, -3, -4, -5)
print(mymodel.correlation)
```
